# Загрузка Pandas и очистка данных

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('main_task.xls')
```

```python
# Ваш код по очистке данных и генерации новых признаков
# При необходимости добавьте ячейки
```

```python
data2 = df
###getting rid of NA values
data2 = data2.dropna()
```

```python
### Now let's Rid of useless columns
data2 = data2.drop(columns= ['URL_TA','ID_TA','Restaurant_id'])
# rid of useless columns
```

### Разбираемся с кухнями

```python
'''### splitting cuisines into words
data2['Cuisine Style'] = data2['Cuisine Style'].apply(lambda x: x.strip("',[]"))
data2['Cuisine Style'] = data2['Cuisine Style'].apply(lambda x: x.split("', '"))'''
```

```python
### list of cuisines
cuisines = set()
for i in data2['Cuisine Style']:
    cuisines.update(i)
print(cuisines)
```

```python
'''## exploding and encoding the cuisines
data2 = data2.explode('Cuisine Style')
dum_cuis = pd.get_dummies(data2['Cuisine Style'])
# merge with main df bridge_df on key values
data2 = data2.join(dum_cuis)
data2.drop(columns='Cuisine Style', inplace = True)'''
```

### Разбираемся с ценником

```python
## make price range numerical
data2['Price Range'].unique()
def dollar_to_number(x):
    if x == '$':
        return 1
    if x == '$$-$$$':
        return 2
    if x == '$$$$':
        return 3
    else:
        return 1
data2['Price Range'] = data2['Price Range'].apply(dollar_to_number)
```

## Splitting cities into dummies

```python
# generate binary values using get_dummies
dum_df = pd.get_dummies(data2.City)
# merge with main df bridge_df on key values
data2 = data2.join(dum_df)
```

```python
### drop cities
data2.drop(columns='City', inplace = True)
```

```python
data2
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = data2.drop(columns=['Reviews','Rating','Cuisine Style'])
y = data2['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
